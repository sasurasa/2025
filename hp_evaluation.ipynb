```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

```

```python
rec_model = pd.read_csv('/content/model_1_final.csv')
rec_model.head()
```

```python
len(rec_model)
```

```python
def get_output(input_value):
  row = rec_model[rec_model['input'].str.contains(input_value)]

  if not row.empty:
    return row['output'].iloc[0]
  else:
    return None
```

```python
validation = pd.read_csv('/content/hp_val_dedup.csv', header=None)
validation = validation.iloc[:, 0:20]
validation.head()

```

```python
len(validation)
```

```python
def process_row(row):
    first_col_model_output = get_output(row[0])
    if first_col_model_output is None:
        first_col_values = []  # Empty list as default
    else:
        first_col_values = [item.strip() for item in first_col_model_output.split(',')]
    for i in range(1, len(row)):
        if pd.isna(row[i]):
            row[i] = 0
        else:
            row[i] = 1 if row[i] in first_col_values else 0
    return row
```

```python
validation_score = validation.apply(process_row, axis=1)
validation_score.head()

```

```python
validation_score['sum'] = validation_score.iloc[:, 1:10].sum(axis=1)
validation_score.head()
```

```python
percent_correct = 100*(len(validation_score[validation_score['sum'] > 0])/len(validation_score))
print(percent_correct)
```
